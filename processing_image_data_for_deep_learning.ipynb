## Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('mikhail-vasilyev-MEb2jandkbc-unsplash.jpg')
img_resized = img.resize((200, 200))
```

```python
img_resized.save('cat_image_resized.jpg')
```

```python
# displaying image using the numpy array
img = mpimg.imread('cat_image_resized.jpg')
img_plot = plt.imshow(img)
plt.show()
```

```python
img.shape
```

## RGB to Grayscale conversion using OpenCV

```python
import cv2
```

```python
img = cv2.imread('cat_image_resized.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscaled_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscaled_image)
```

```python
grayscaled_image.shape
```

```python
# displaying image using the numpy array
plt.imshow(grayscaled_image, cmap='gray')
plt.show()
```

```python
# Saving the grayscaled image
cv2.imwrite('grayscaled_image_cat.jpg', grayscaled_image)
```
